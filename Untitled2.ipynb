```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import scipy
```

```python
soybean_oil = pd.read_csv("data/Zl Soybean Oil.csv",skiprows=range(1, 501))
soybean_meal = pd.read_csv("data/Zm Soybean Meal.csv", skiprows=range(1, 501))
soybean_mini = pd.read_csv("data/Zs Soybean Mini.csv", skiprows=range(1, 501))
soybean_oil

```

```python
soybean_spread_open = (11 * soybean_oil['open']) + (48 * soybean_meal['open']) - (50 * soybean_mini['open'])
soybean_spread_high = (11 * soybean_oil['high']) + (48 * soybean_meal['high']) - (50 * soybean_mini['low'])
soybean_spread_low = (11 * soybean_oil['low']) + (48 * soybean_meal['low']) - (50 * soybean_mini['high'])
soybean_spread_close = (11 * soybean_oil['close']) + (48 * soybean_meal['close']) - (50 * soybean_mini['close'])
soybean_spread_volume = (soybean_oil['volume'] + soybean_meal['volume'] + soybean_mini['volume']) // 3
```

```python
def fractional_difference(series, alpha, window_size):
    series = np.array(series)


    diff_series = np.zeros_like(series)


    for t in range(window_size, len(series)):
        diff = 0
        for k in range(window_size):

            binomial_coeff = scipy.special.comb(alpha, k)

            diff += (-1) ** k * binomial_coeff * series[t - k - 1]


        diff_series[t] = diff

    return diff_series

differentiated_spread = fractional_difference(soybean_spread_close, 1, 3)
differentiated_spread_volume = fractional_difference(soybean_spread_volume, 1, 3)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(differentiated_spread, linestyle='-', color='b', label='Differentiated Data')
plt.title('Differentiated Data (diff_manual)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()
```

```python

differentiated_spread = differentiated_spread[~np.isnan(differentiated_spread)]
differentiated_spread_volume = differentiated_spread_volume[~np.isnan(differentiated_spread_volume)]
# 1. Cyclical Component
decomposition = seasonal_decompose(differentiated_spread, model='additive', period=50)
cyclical_component = decomposition.resid

# 2. Rolling Volatility
rolling_volatility = pd.Series(differentiated_spread).rolling(window=20).std().values

# 3. Seasonal Pattern
seasonal_sin = np.sin(2 * np.pi * np.arange(len(differentiated_spread)) / 365)
seasonal_cos = np.cos(2 * np.pi * np.arange(len(differentiated_spread)) / 365)

# 4. Normalized Volume-Weighted
volume_weighted = (differentiated_spread * differentiated_spread_volume) / differentiated_spread_volume.sum()
norm_volume_weighted = (volume_weighted - volume_weighted.mean()) / volume_weighted.std()
```

```python
import xgboost as xgb
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, RobustScaler
import optuna
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose


decomp_50 = seasonal_decompose(differentiated_spread, model='additive', period=50)
decomp_20 = seasonal_decompose(differentiated_spread, model='additive', period=20)
cyclical_component_50 = decomp_50.resid
cyclical_component_20 = decomp_20.resid


rolling_vol_20 = pd.Series(differentiated_spread).rolling(window=20).std()
rolling_vol_50 = pd.Series(differentiated_spread).rolling(window=50).std()
ewm_vol = pd.Series(differentiated_spread).ewm(span=20).std()


seasonal_sin_annual = np.sin(2 * np.pi * np.arange(len(differentiated_spread)) / 365)
seasonal_cos_annual = np.cos(2 * np.pi * np.arange(len(differentiated_spread)) / 365)
seasonal_sin_quarter = np.sin(2 * np.pi * np.arange(len(differentiated_spread)) / 90)
seasonal_cos_quarter = np.cos(2 * np.pi * np.arange(len(differentiated_spread)) / 90)


volume_weighted = (differentiated_spread * differentiated_spread_volume) / differentiated_spread_volume.sum()
norm_volume_weighted = (volume_weighted - volume_weighted.mean()) / volume_weighted.std()
vol_price_ratio = pd.Series(differentiated_spread_volume).rolling(window=20).mean() / rolling_vol_20


momentum_20 = pd.Series(differentiated_spread).diff(20)
momentum_50 = pd.Series(differentiated_spread).diff(50)


zscore_20 = (differentiated_spread - pd.Series(differentiated_spread).rolling(window=20).mean()) / rolling_vol_20
zscore_50 = (differentiated_spread - pd.Series(differentiated_spread).rolling(window=50).mean()) / rolling_vol_50


features = np.column_stack([
    cyclical_component_20, cyclical_component_50,
    rolling_vol_20, rolling_vol_50, ewm_vol,
    seasonal_sin_annual, seasonal_cos_annual,
    seasonal_sin_quarter, seasonal_cos_quarter,
    norm_volume_weighted, vol_price_ratio,
    momentum_20, momentum_50,
    zscore_20, zscore_50
])


features = pd.DataFrame(features).fillna(method='ffill').fillna(method='bfill').values
scaler = RobustScaler()
scaled_features = scaler.fit_transform(features)


target = np.sign(np.roll(differentiated_spread, -5) - differentiated_spread)
target = target[:-5]
target = (target + 1) / 2


min_length = min(len(scaled_features), len(target))
scaled_features = scaled_features[:min_length]
target = target[:min_length]

def prepare_data(features, target, lookback=10):  # Increased lookback
    X, y = [], []
    for i in range(len(features) - lookback):
        if i + lookback < len(target):
            X.append(features[i:i+lookback])
            y.append(target[i+lookback])
    return np.array(X), np.array(y)


X, y = prepare_data(scaled_features, target, lookback=10)
X_reshaped = X.reshape(X.shape[0], -1)


def objective(trial):
    params = {
        'objective': 'binary:logistic',
        'max_depth': trial.suggest_int('max_depth', 4, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.05, log=True),
        'n_estimators': trial.suggest_int('n_estimators', 200, 2000),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'subsample': trial.suggest_float('subsample', 0.5, 0.9),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 0.9),
        'gamma': trial.suggest_float('gamma', 0, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 2.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 2.0),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 0.8, 1.2)
    }

    tscv = TimeSeriesSplit(n_splits=5, gap=5)
    sharpes = []

    for train_idx, val_idx in tscv.split(X_reshaped):
        X_train_cv, X_val_cv = X_reshaped[train_idx], X_reshaped[val_idx]
        y_train_cv, y_val_cv = y[train_idx], y[val_idx]

        model = xgb.XGBClassifier(**params, random_state=42)
        model.fit(X_train_cv, y_train_cv,
                 eval_set=[(X_val_cv, y_val_cv)],
                 verbose=False)

        proba = model.predict_proba(X_val_cv)[:, 1]
        returns = (2 * y_val_cv - 1) * proba
        sharpe = np.sqrt(252) * returns.mean() / returns.std()
        sharpes.append(sharpe)

    return -np.mean(sharpes)


study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100)


best_params = study.best_params
best_params['objective'] = 'binary:logistic'
final_model = xgb.XGBClassifier(**best_params, random_state=42)

split_index = int(0.8 * len(X_reshaped))

X_train = X_reshaped[:split_index]
y_train = y[:split_index]
X_test = X_reshaped[split_index:]
y_test = y[split_index:]

final_model.fit(X_train, y_train,
                eval_set=[(X_train, y_train), (X_test, y_test)])


train_pred = final_model.predict(X_train)
test_pred = final_model.predict(X_test)
train_prob = final_model.predict_proba(X_train)
test_prob = final_model.predict_proba(X_test)

def print_metrics(y_true, y_pred, proba, set_name):

    y_true_orig = 2 * y_true - 1
    y_pred_orig = 2 * y_pred - 1

    print(f"\n{set_name} Metrics:")
    print(f"MSE: {mean_squared_error(y_true_orig, y_pred_orig):.4f}")
    print(f"MAE: {mean_absolute_error(y_true_orig, y_pred_orig):.4f}")
    print(f"R2: {r2_score(y_true_orig, y_pred_orig):.4f}")

    returns = y_true_orig * proba[:, 1]
    sharpe = np.sqrt(252) * returns.mean() / returns.std()
    print(f"Sharpe Ratio: {sharpe:.4f}")

    cm = pd.crosstab(y_true, y_pred, margins=True)
    print("\nConfusion Matrix:")
    print(cm)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{set_name} Confusion Matrix')
    plt.show()


print_metrics(y_train, train_pred, train_prob, "Training")
print_metrics(y_test, test_pred, test_prob, "Testing")

plt.figure(figsize=(10, 6))
xgb.plot_importance(final_model)
plt.title('Feature Importance')
plt.show()
```
